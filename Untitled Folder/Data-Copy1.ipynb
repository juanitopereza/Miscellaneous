## Reading data

```python
#Reading data
# Path
path = './dataWind/'# data directory
# import datafile
txt05 = pd.read_csv(path+'wind005.txt',        # path of the file to read
                   header=None,        # remove header if any
                   #sep="\s+",          # separator
                   names=["z5"])  # give name to columns
txt15 = pd.read_csv(path+'wind015.txt',        # path of the file to read
                   header=None,        # remove header if any
                   names=["z15"])  # give name to columns
txt25 = pd.read_csv(path+'wind025.txt',        # path of the file to read
                   header=None,        # remove header if any
                   names=["z25"])  # give name to columns
txt35 = pd.read_csv(path+'wind035.txt',        # path of the file to read
                   header=None,        # remove header if any
                   names=["z35"])  # give name to columns
txt45 = pd.read_csv(path+'wind045.txt',        # path of the file to read
                   header=None,        # remove header if any
                   names=["z45"])  # give name to columns
txt55 = pd.read_csv(path+'wind055.txt',        # path of the file to read
                   header=None,        # remove header if any
                   names=["z55"])  # give name to columns
txt65 = pd.read_csv(path+'wind065.txt',        # path of the file to read
                   header=None,        # remove header if any
                   names=["z65"])  # give name to columns
txt75 = pd.read_csv(path+'wind075.txt',        # path of the file to read
                   header=None,        # remove header if any
                   names=["z75"])  # give name to columns
txt85 = pd.read_csv(path+'wind085.txt',        # path of the file to read
                   header=None,        # remove header if any
                   names=["z85"])  # give name to columns
txt95 = pd.read_csv(path+'wind095.txt',        # path of the file to read
                   header=None,        # remove header if any
                   names=["z95"])  # give name to columns
txt105 = pd.read_csv(path+'wind105.txt',        # path of the file to read
                   header=None,        # remove header if any
                   names=["z105"])  # give name to columns
txt115 = pd.read_csv(path+'wind115.txt',        # path of the file to read
                   header=None,        # remove header if any
                   names=["z115"])  # give name to columns

data = pd.concat([txt05,txt15,txt25,txt35,txt45,txt55,txt65,txt75,txt85,txt95,txt105,txt115], axis=1)
```

## Calculation of the velocity

```python
coefa=4.63171324459
coefb=1.48813234443
coefn=0.641866655328
U1= ((((txt05**2)-coefa))/coefb)**(1/coefn)
U2= ((((txt15**2)-coefa))/coefb)**(1/coefn)
U3= ((((txt25**2)-coefa))/coefb)**(1/coefn)
U4= ((((txt35**2)-coefa))/coefb)**(1/coefn)
U5= ((((txt45**2)-coefa))/coefb)**(1/coefn)
U6= ((((txt55**2)-coefa))/coefb)**(1/coefn)
U7= ((((txt65**2)-coefa))/coefb)**(1/coefn)
U8= ((((txt75**2)-coefa))/coefb)**(1/coefn)
U9= ((((txt85**2)-coefa))/coefb)**(1/coefn)
U10= ((((txt95**2)-coefa))/coefb)**(1/coefn)
U11= ((((txt105**2)-coefa))/coefb)**(1/coefn)
U12= ((((txt115**2)-coefa))/coefb)**(1/coefn)
U = pd.concat([U1,U2,U3,U4,U5,U6,U7,U8,U9,U10,U11,U12], axis=1)
```

```python
# number of data points:
n=len(U)
#time of measurements ins seconds
t=60
#sampling frecuency 1000 hz
# generation of timestamps
U.index = pd.date_range('1/1/2000',    # start of the time stamp
                      periods=n,          # number of points
                      freq='1L')        #frequency (here 1000 milliseconds)
# Other frequencies:
# L milliseconds
# U microseconds
# S seconds
# T minuts
# H hours
# D day
# W week
# M month
# Y year
```

```python
# Computation of the mean 
# use function: data.mean()
mean = U.mean()
# Computation of the standard deviation
# use function: data.std()
std = U.std()
# Skewness
# use function: data.skew()
skew = U.skew()
# Kurtosis: must be equal to 3 for a Gaussian process
# use function: data.kurtosis() +3
kurt = U.kurtosis() +3
print(kurt)
```

## computing fluctuations

```python
# for computing the fluctuations we need to remember that u'= u - u(mean)
#removing the mean
z5prime= U.z5 - mean.z5
z15prime= U.z15 - mean.z15
z25prime= U.z25 - mean.z25
z35prime= U.z35 - mean.z35
z45prime= U.z45 - mean.z45
z55prime= U.z55 - mean.z55
z65prime= U.z65 - mean.z65
z75prime= U.z75 - mean.z75
z85prime= U.z85 - mean.z85
z95prime= U.z95 - mean.z95
z105prime= U.z105 - mean.z105
z115prime= U.z115 - mean.z115
```

##  computing Spectra

```python
# computing Spectra using 
# f,spec = signal.welch(data, fs=fsamp, window='hann', nperseg=npseg, noverlap=npseg*overl, nfft=npseg, detrend=False, return_onesided=True, scaling='density', axis=-1)
# output are f (frequency grid) and spec (array containing the corresponding PSD values as a function of f)
#npseg is the number of blocs that we want to divide the signal
fsamp = 1000.
dt = 1./fsamp
# Parameters for Welch
#npseg = 2.**15
npseg = 1.65**15
overl = .5
nfft = npseg
df = fsamp/nfft
# PSD of u1 and u2
fz5,specu1 = signal.welch(z5prime, fs=fsamp, window='hann', nperseg=npseg, noverlap=npseg*overl, nfft=npseg, detrend=False, return_onesided=True, scaling='density', axis=-1)
fz15,specu2 = signal.welch(z15prime, fs=fsamp, window='hann', nperseg=npseg, noverlap=npseg*overl, nfft=npseg, detrend=False, return_onesided=True, scaling='density', axis=-1)
fz25,specu3 = signal.welch(z25prime, fs=fsamp, window='hann', nperseg=npseg, noverlap=npseg*overl, nfft=npseg, detrend=False, return_onesided=True, scaling='density', axis=-1)
fz35,specu4 = signal.welch(z35prime, fs=fsamp, window='hann', nperseg=npseg, noverlap=npseg*overl, nfft=npseg, detrend=False, return_onesided=True, scaling='density', axis=-1)
fz45,specu5 = signal.welch(z45prime, fs=fsamp, window='hann', nperseg=npseg, noverlap=npseg*overl, nfft=npseg, detrend=False, return_onesided=True, scaling='density', axis=-1)
fz55,specu6 = signal.welch(z55prime, fs=fsamp, window='hann', nperseg=npseg, noverlap=npseg*overl, nfft=npseg, detrend=False, return_onesided=True, scaling='density', axis=-1)
fz65,specu7 = signal.welch(z65prime, fs=fsamp, window='hann', nperseg=npseg, noverlap=npseg*overl, nfft=npseg, detrend=False, return_onesided=True, scaling='density', axis=-1)
fz75,specu8 = signal.welch(z75prime, fs=fsamp, window='hann', nperseg=npseg, noverlap=npseg*overl, nfft=npseg, detrend=False, return_onesided=True, scaling='density', axis=-1)
fz85,specu9 = signal.welch(z85prime, fs=fsamp, window='hann', nperseg=npseg, noverlap=npseg*overl, nfft=npseg, detrend=False, return_onesided=True, scaling='density', axis=-1)
fz95,specu10 = signal.welch(z95prime, fs=fsamp, window='hann', nperseg=npseg, noverlap=npseg*overl, nfft=npseg, detrend=False, return_onesided=True, scaling='density', axis=-1)
fz105,specu11 = signal.welch(z105prime, fs=fsamp, window='hann', nperseg=npseg, noverlap=npseg*overl, nfft=npseg, detrend=False, return_onesided=True, scaling='density', axis=-1)
fz115,specu12 = signal.welch(z115prime, fs=fsamp, window='hann', nperseg=npseg, noverlap=npseg*overl, nfft=npseg, detrend=False, return_onesided=True, scaling='density', axis=-1)

# Variance check: integral of the PSD = signal variance
#singal variance
varu1= (std.z5)**2
varu2= (std.z15)**2
varu3= (std.z25)**2
varu4= (std.z35)**2
varu5= (std.z45)**2
varu6= (std.z55)**2
varu7= (std.z65)**2
varu8= (std.z75)**2
varu9= (std.z85)**2
varu10= (std.z95)**2
varu11= (std.z105)**2
varu12= (std.z115)**2
#integral of PSD
PSDintu1= (np.sum(specu1)) * df
PSDintu2= (np.sum(specu2)) * df
PSDintu3= (np.sum(specu3)) * df
PSDintu4= (np.sum(specu4)) * df
PSDintu5= (np.sum(specu5)) * df
PSDintu6= (np.sum(specu6)) * df
PSDintu7= (np.sum(specu7)) * df
PSDintu8= (np.sum(specu8)) * df
PSDintu9= (np.sum(specu9)) * df
PSDintu10= (np.sum(specu10)) * df
PSDintu11= (np.sum(specu11)) * df
PSDintu12= (np.sum(specu12)) * df
#Display
print('Check U1:',varu1,PSDintu1)
print('Check U2:',varu2,PSDintu2)
print('Check U3:',varu3,PSDintu3)
print('Check U4:',varu4,PSDintu4)
print('Check U5:',varu5,PSDintu5)
print('Check U6:',varu6,PSDintu6)
print('Check U7:',varu7,PSDintu7)
print('Check U8:',varu8,PSDintu8)
print('Check U9:',varu9,PSDintu9)
print('Check U10:',varu10,PSDintu10)
print('Check U11:',varu11,PSDintu11)
print('Check U12:',varu12,PSDintu12)
```

```python
# Plotting in log-log coordinate
plt.subplot(211)
plt.loglog(fz5,specu1 ,'g--', fz15,specu2, 'm--',fz25,specu3,fz35,specu4,fz45,specu5,fz55,specu6)
plt.xlabel('power spectral density ')
plt.ylabel('frequency grid')
plt.legend(('z5','z15','z25','z35','z45','z55'))
plt.subplot(212)
plt.loglog(fz65,specu7,fz75,specu8,fz85,specu9,fz95,specu10,fz105,specu11,fz115,specu12)
plt.xlabel('power spectral density ')
plt.ylabel('frequency grid')
plt.legend(('z65','z75','z85','z95','z105','z115'))
plt.show()
```

### Presence of white noise

```python
# Superimposition of white (random) noise onto the signal
# Noise
nsig = (.25)**.5 # standard deviation of noise
noise = np.random.randn(n)*nsig
# PSD of noise
fn,specun = signal.welch(noise, fs=fsamp, window='hann', nperseg=npseg, noverlap=npseg*overl, nfft=npseg, detrend=False, return_onesided=True, scaling='density', axis=-1)
# PSD of data + noise
datanu1= z5prime + noise
fu1n,specu1n = signal.welch(datanu1, fs=fsamp, window='hann', nperseg=npseg, noverlap=npseg*overl, nfft=npseg, detrend=False, return_onesided=True, scaling='density', axis=-1)
datanu12= z115prime + noise
fu12n,specu12n = signal.welch(datanu12, fs=fsamp, window='hann', nperseg=npseg, noverlap=npseg*overl, nfft=npseg, detrend=False, return_onesided=True, scaling='density', axis=-1)
#
# Variance check: integral of the PSD = signal variance
varu1n= ((std.z5)**2)+(nsig**2)
varu12n= ((std.z115)**2)+(nsig**2)
PSDintu1n= (np.sum(specu1n)) * df
PSDintu12n= (np.sum(specu12n)) * df
#Display
print('Check U1:',varu1n,PSDintu1n,'Check U12:',varu12n,PSDintu12n)
```
