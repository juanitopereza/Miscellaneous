## Reading data

```python
#Reading data
# Path
path = './dataWind/'# data directory
# import datafile
txt05 = pd.read_csv(path+'wind005.txt',        # path of the file to read
                   header=None,        # remove header if any
                   #sep="\s+",          # separator
                   names=["z5"])  # give name to columns
txt15 = pd.read_csv(path+'wind015.txt',        # path of the file to read
                   header=None,        # remove header if any
                   names=["z15"])  # give name to columns
txt25 = pd.read_csv(path+'wind025.txt',        # path of the file to read
                   header=None,        # remove header if any
                   names=["z25"])  # give name to columns
txt35 = pd.read_csv(path+'wind035.txt',        # path of the file to read
                   header=None,        # remove header if any
                   names=["z35"])  # give name to columns
txt45 = pd.read_csv(path+'wind045.txt',        # path of the file to read
                   header=None,        # remove header if any
                   names=["z45"])  # give name to columns
txt55 = pd.read_csv(path+'wind055.txt',        # path of the file to read
                   header=None,        # remove header if any
                   names=["z55"])  # give name to columns
txt65 = pd.read_csv(path+'wind065.txt',        # path of the file to read
                   header=None,        # remove header if any
                   names=["z65"])  # give name to columns
txt75 = pd.read_csv(path+'wind075.txt',        # path of the file to read
                   header=None,        # remove header if any
                   names=["z75"])  # give name to columns
txt85 = pd.read_csv(path+'wind085.txt',        # path of the file to read
                   header=None,        # remove header if any
                   names=["z85"])  # give name to columns
txt95 = pd.read_csv(path+'wind095.txt',        # path of the file to read
                   header=None,        # remove header if any
                   names=["z95"])  # give name to columns
txt105 = pd.read_csv(path+'wind105.txt',        # path of the file to read
                   header=None,        # remove header if any
                   names=["z105"])  # give name to columns
txt115 = pd.read_csv(path+'wind115.txt',        # path of the file to read
                   header=None,        # remove header if any
                   names=["z115"])  # give name to columns

data = pd.concat([txt05,txt15,txt25,txt35,txt45,txt55,txt65,txt75,txt85,txt95,txt105,txt115], axis=1)
```

## Calculation of the velocity

```python
coefa=4.63171324459
coefb=1.48813234443
coefn=0.641866655328
U1= ((((txt05**2)-coefa))/coefb)**(1/coefn)
U2= ((((txt15**2)-coefa))/coefb)**(1/coefn)
U3= ((((txt25**2)-coefa))/coefb)**(1/coefn)
U4= ((((txt35**2)-coefa))/coefb)**(1/coefn)
U5= ((((txt45**2)-coefa))/coefb)**(1/coefn)
U6= ((((txt55**2)-coefa))/coefb)**(1/coefn)
U7= ((((txt65**2)-coefa))/coefb)**(1/coefn)
U8= ((((txt75**2)-coefa))/coefb)**(1/coefn)
U9= ((((txt85**2)-coefa))/coefb)**(1/coefn)
U10= ((((txt95**2)-coefa))/coefb)**(1/coefn)
U11= ((((txt105**2)-coefa))/coefb)**(1/coefn)
U12= ((((txt115**2)-coefa))/coefb)**(1/coefn)
U = pd.concat([U1,U2,U3,U4,U5,U6,U7,U8,U9,U10,U11,U12], axis=1)
```

```python
# number of data points:
n=len(U)
#time of measurements ins seconds
t=60
#sampling frecuency 1000 hz
# generation of timestamps
U.index = pd.date_range('1/1/2000',    # start of the time stamp
                      periods=n,          # number of points
                      freq='1L')        #frequency (here 1000 milliseconds)
# Other frequencies:
# L milliseconds
# U microseconds
# S seconds
# T minuts
# H hours
# D day
# W week
# M month
# Y year
```

```python
# Computation of the mean 
# use function: data.mean()
mean = U.mean()
# Computation of the standard deviation
# use function: data.std()
std = U.std()
# Skewness
# use function: data.skew()
skew = U.skew()
# Kurtosis: must be equal to 3 for a Gaussian process
# use function: data.kurtosis() +3
kurt = U.kurtosis() +3
print(kurt)
```

## computing fluctuations

```python
# for computing the fluctuations we need to remember that u'= u - u(mean)
#removing the mean
u1prime= U.z5 - mean.z5
u2prime= U.z15 - mean.z15
u3prime= U.z25 - mean.z25
u4prime= U.z35 - mean.z35
u5prime= U.z45 - mean.z45
u6prime= U.z55 - mean.z55
u7prime= U.z65 - mean.z65
u8prime= U.z75 - mean.z75
u9prime= U.z85 - mean.z85
u10prime= U.z95 - mean.z95
u11prime= U.z105 - mean.z105
u12prime= U.z115 - mean.z115
```

### Correlation computation

```python
# compute correlation with the fluctuation with corr = signal.correlate(data,data,mode='full', method='fft')
# Be careful of normalization
corru1 = signal.correlate(u1prime,u1prime,mode='full', method='fft')
corru2= signal.correlate(u2prime,u2prime,mode='full', method='fft')
corru3 = signal.correlate(u3prime,u3prime,mode='full', method='fft')
corru4= signal.correlate(u4prime,u4prime,mode='full', method='fft')
corru5 = signal.correlate(u5prime,u5prime,mode='full', method='fft')
corru6= signal.correlate(u6prime,u6prime,mode='full', method='fft')
corru7 = signal.correlate(u7prime,u7prime,mode='full', method='fft')
corru8= signal.correlate(u8prime,u8prime,mode='full', method='fft')
corru9 = signal.correlate(u9prime,u9prime,mode='full', method='fft')
corru10= signal.correlate(u10prime,u10prime,mode='full', method='fft')
corru11 = signal.correlate(u11prime,u11prime,mode='full', method='fft')
corru12= signal.correlate(u12prime,u12prime,mode='full', method='fft')
#Because python gives an "incomplete" correlation we need to find the correlation of a vector of 1 with the size of data
#creating the vector array full of ones where n is the size of the array
#ones= np.array(n*[1], dtype='f') also works
ones= np.ones(n,dtype='f')
#finding the correlation of the ones vector
corr = signal.correlate(ones,ones,mode='full', method='fft')
#finding the real correlation
corru1=corru1/corr
corru2=corru2/corr
corru3=corru3/corr
corru4=corru4/corr
corru5=corru5/corr
corru6=corru6/corr
corru7=corru7/corr
corru8=corru8/corr
corru9=corru9/corr
corru10=corru10/corr
corru11=corru11/corr
corru12=corru12/corr
#correlation coeficient: will be the correlation (R) divided in de standard derivation
ccoefu1=corru1/((std.z5)**2)
ccoefu2=corru2/((std.z15)**2)
ccoefu3=corru3/((std.z25)**2)
ccoefu4=corru4/((std.z35)**2)
ccoefu5=corru5/((std.z45)**2)
ccoefu6=corru6/((std.z55)**2)
ccoefu7=corru7/((std.z65)**2)
ccoefu8=corru8/((std.z75)**2)
ccoefu9=corru9/((std.z85)**2)
ccoefu10=corru10/((std.z95)**2)
ccoefu11=corru11/((std.z105)**2)
ccoefu12=corru12/((std.z115)**2)
# compute lagstime-
dt = 1/(2*t)
lags = np.arange(-n + 1, n) # so last value is nx - 1
tlags = lags*dt
```
